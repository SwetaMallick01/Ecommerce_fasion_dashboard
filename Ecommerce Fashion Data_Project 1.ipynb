```python
import pandas as pd
import numpy as np
```

```python
fashion = pd.read_csv(r"C:\Users\sweta\Desktop\Jupyter Practice\FashionDataset.csv")
```

```python
fashion.head()
```

```python
fashion.isnull().sum()
fashion.info()
```

```python
fashion.drop("Unnamed: 0", axis = 1, inplace= True)
fashion.drop("Sizes", axis = 1, inplace= True)
```

```python
fashion.head()
```

```python
fashion["MRP"] = fashion["MRP"].str.replace("Rs\n", "")
```

```python
fashion["MRP"] = fashion["MRP"].replace("NaN", np.NaN)
fashion["MRP"] = fashion["MRP"].replace("Nan", np.NaN)

fashion["MRP"] = fashion["MRP"].dropna(axis = 0)
fashion = fashion.dropna(axis = 0)
```

```python
fashion.head()
```

```python
fashion["Discount"] = fashion["Discount"].str.replace("% off", "").astype(int)
fashion["Category"] = fashion["Category"].str.replace("-Women", "")
```

```python
fashion.head()
```

```python
fashion['MRP'] = fashion['MRP'].astype(int)
```

```python
fashion['Discount_value'] = fashion['MRP']*0.01*fashion['Discount']
```
